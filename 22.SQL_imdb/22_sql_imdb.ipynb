### (Q1) List all the directors who directed a 'Comedy' movie in a leap year. (You need to check that the genre is 'Comedy’ and year is a leap year) Your query should return director name, the movie name, and the year.

```python
query = """
    SELECT DISTINCT p.name, m.title, m.year FROM person p, movie m JOIN 
    (SELECT md.pid AS pid, mg.mid AS mid FROM m_director md JOIN m_genre mg ON mg.mid=md.mid
    WHERE mg.gid IN
    (SELECT g.gid FROM genre g WHERE g.name LIKE "%Comedy%")) AS temp ON temp.pid=p.pid AND temp.mid=m.mid group by m.title
    ;
"""
result = pd.read_sql_query(query, conn)
result
```

```python
query = """
    SELECT DISTINCT p.name, m.title FROM person p, movie m WHERE (p.pid, m.mid) IN
    (SELECT md.pid AS pid, mg.mid AS mid FROM m_director md JOIN m_genre mg ON mg.mid=md.mid
    WHERE mg.gid IN
    (SELECT g.gid FROM genre g WHERE g.name LIKE "%Comedy%")) 
    ;
"""
result = pd.read_sql_query(query, conn)
result
```

### (Q3) List all the actors who acted in a film before 1970 and in a film after 1990. (That is: < 1970 and > 1990.)

```python
query = """
select p.name, m.year from person p, movie m , m_cast mc
where p.pid=mc.pid and mc.mid=m.mid
and m.year not between 1970 and 1990
"""
result = pd.read_sql_query(query, conn)
result
```
